# Exploratory Data Analysis of Netflix Dataset

### About Netflix

Netflix is a leading global streaming platform that revolutionized the way people consume entertainment. Founded in 1997 by Reed Hastings and Marc Randolph, the company started as a DVD rental service by mail. However, it quickly evolved into a subscription-based online streaming service, launching its digital platform in 2007. This strategic move allowed users to access a vast library of movies, TV shows, documentaries, and original content on various devices.

With its headquarters in Los Gatos, California, Netflix's popularity soared due to its user-friendly interface, affordable pricing, and personalized content recommendations. By leveraging sophisticated algorithms, the platform tailored viewing suggestions based on users' preferences, keeping them engaged and entertained.

Netflix's original content strategy played a pivotal role in its success, producing award-winning series like "House of Cards," "Stranger Things," and "The Crown." This emphasis on exclusive content attracted a growing subscriber base, surpassing 200 million global subscribers by 2021.

The company's global expansion was another key factor in its triumph, as it successfully penetrated international markets, offering a localized experience with subtitles and dubbing in multiple languages. Additionally, it embraced diverse cultural perspectives, producing content that resonated with audiences worldwide.

As of 2021, Netflix faced increased competition from other streaming services like Amazon Prime Video, Disney+, and Hulu, leading to a more fragmented market. Despite the challenges, Netflix's innovation, data-driven approach, and dedication to high-quality entertainment have solidified its position as a media giant and a dominant force in the streaming industry.

### Handling Missing Values

rating,cast,country and director have the nulls filled with 'Unavailable'

```python
df.fillna({'rating': 'Unavailable','cast':'Unavailable','country':'Unavailable','director':'Unavailable'},inplace=True)
df.isna().sum()
```

For nulls in date_added, missing date_added can be subsituted with the most recent date from date_added. This is because Netflix
has the tendency to add more content over time.

```python
df[df.date_added.isnull()]
```

```python
most_recent_entry_date=df['date_added'].max()
df.fillna({'date_added': most_recent_entry_date}, inplace = True)
```

```python
df[df.show_id == 's8183']
```

#### Additional Data Cleansing

```python
df[df.duration.isnull()]
```

The missing durations are all movies by Louis C.K. Normally, we would likely fill the duration with the mean duration of movies
from the table.In this case it appears that the actual duration was input into the rating column, so one solution is to move the
rating data into the duration and the make the rating information 'Unavailable' like the other nulls.

Check to make sure that there is no other content with the same director to avoid accidental overwriting.

```python
df[df.director == 'Louis C.K.'].head()
```

Overwrite and Check

```python
# loc helps us accessing the columns by names
df.loc[df['director'] == 'Louis C.K.', 'duration'] = df['rating']
df[df.director == 'Louis C.K.'].head()
```

Second Overwrite and Check

```python
df.loc[df['director'] == 'Louis C.K.', 'rating'] = 'Unavailable'
df[df.director == 'Louis C.K.'].head()
```
